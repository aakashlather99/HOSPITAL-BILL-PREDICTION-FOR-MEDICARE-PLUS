```python
#getting info about columns so that we can define x an y variable
df.columns
```

```python
# define y and x for testing and training
y=df['predicted_hospital_bill_inr']
x=df[['gender', 'age', 'insurance_type', 'diagnosis_type', 'procedures_count',
       'surgery_done', 'icu_admission', 'length_of_stay_days',
       'medications_count', 'room_type']]
```

```python
#split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
#select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#train the model
# Ensure 'insurance_type' is numerical before training
x_train['insurance_type'] = x_train['insurance_type'].replace({'Government': 0, 'Private': 1}).astype(int)

model.fit(x_train,y_train)
```

```python
#intercept and coefficients
model.intercept_,model.coef_
```

```python
#prediction
# Ensure 'insurance_type' is numerical before prediction
x_test['insurance_type'] = x_test['insurance_type'].replace({'Government': 0, 'Private': 1}).astype(int)

y_pred=model.predict(x_test)
```

```python
#evaluate
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

## Project Introduction: Hospital Bill Prediction for MediCarePlus

This project focuses on developing a machine learning model to predict hospital bills for MediCarePlus patients. The goal is to leverage historical patient data to build a predictive tool that can estimate the cost of hospitalization based on various factors.

The dataset used in this project includes information such as patient demographics (gender, age), insurance type, diagnosis, the number of procedures and medications, whether surgery was performed or ICU admission was required, the length of hospital stay, and the type of room occupied.

By analyzing the relationships between these features and the final hospital bill amount, the machine learning model aims to provide accurate cost predictions. This can be valuable for various applications, including financial planning for both patients and the healthcare provider, optimizing resource allocation, and potentially identifying factors that significantly influence hospital costs. This is a regression problem, as the model is predicting a continuous numerical value (the hospital bill amount).
